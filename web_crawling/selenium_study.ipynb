```python
# 크롬 드라이버
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
driver.get("https://www.naver.com")

time.sleep(3)

# 원격으로 검색하기
searchValue = input("검색어 입력: ")

searchInput = driver.find_element(by=By.CSS_SELECTOR, value="#query")
searchInput.send_keys(searchValue)              # input을 통해 입력한 값이 검색창에 입력됨
searchInput.send_keys(Keys.RETURN)              # 자동으로 enter가 되어 검색된 화면으로 이동

time.sleep(2)       #렌더링된 이후에 코드를 실행해야 하므로 time설정은 중요함!

driver.find_element(by=By.XPATH, value='//*[@id="lnb"]/div[1]/div/div[1]/div/div[2]/div[2]/a').click()  # 오른쪽으로 이동버튼 클릭
time.sleep(1)
driver.find_element(by=By.XPATH, value='//*[@id="lnb"]/div[1]/div/div[1]/div/div[1]/div[7]/a').click()  # 뉴스탭 클릭
time.sleep(2)

# 뉴스 타이틀 가져오기
# newsTitles = driver.find_elements(by=By.CSS_SELECTOR, value=".news_tit")
# for title in newsTitles:
#     print(title.text)

for i in range(2, 5):
    driver.find_element(by=By.XPATH, value=f'//*[@id="main_pack"]/div[2]/div/div/a[{i}]').click()
    time.sleep(2)
    # newsTitles = driver.find_elements(by=By.CSS_SELECTOR, value=".news_tit")
    # for title in newsTitles:
    #     print(title.get_attribute(name="href"))
    newsImages = driver.find_elements(by=By.CSS_SELECTOR, value=".dsc_thumb .thumb")
    for img in newsImages:
        print(img.get_attribute(name="src"))

webtoonList = [
    {
        "dayOfTheWeek": "월",
        "webtoon" : [
            {
                "title": "참교육",
                "author":"채용택 / 한가람",
                "grade": 9.87,
                "postImage": "https://"
            }
        ]
    }
]
```
