```python
#  
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import requests
from bs4 import BeautifulSoup as bs

driver = webdriver.Chrome()
driver.get("https://comic.naver.com/index")

time.sleep(3)

driver.find_element(by=By.XPATH, value='//*[@id="menu"]/li[2]/a').click() #  
time.sleep(2)

webtoonList = []
for i in range(2, 9): #  
    day = driver.find_element(by=By.XPATH, value=f'//*[@id="wrap"]/header/div[3]/nav/ul/li[{i}]/a')
    day.click() # 
    time.sleep(3)

    dayOfTheWeek = day.get_attribute("textContent")

    # 
    scrollTop = 0
    while True:
        driver.execute_script(f"window.scrollTo(0, {scrollTop});")

        scrollTop += 500
        time.sleep(0.05)
        if scrollTop > driver.execute_script("return document.body.scrollHeight"):
            break

    webtoons = driver.find_elements(by=By.CSS_SELECTOR, value=".ContentList__content_list--q5KXY .item")
    print(webtoons)
    webtoon_items = []

    for webtoon in webtoons:
        title = webtoon.find_element(by=By.CSS_SELECTOR, value='.ContentTitle__title--e3qXt .text').text
        author = webtoon.find_element(by=By.CSS_SELECTOR, value='.ContentAuthor__author--CTAAP').text
        grade = webtoon.find_element(by=By.CSS_SELECTOR, value='.Rating__star_area--dFzsb .text').text
        postImage = webtoon.find_element(by=By.CSS_SELECTOR, value='.Poster__image--d9XTI').get_attribute(name="src")

        naverweb = {
            "title": title,
            "author":author,
            "grade": grade,
            "postImage": postImage
        }
        webtoon_items.append(naverweb)

    daywebtoon = {
        "dayOfTheWeek": dayOfTheWeek,
        "webtoon": webtoon_items
    }
    webtoonList.append(daywebtoon)
webtoonList
```

```python
import pandas as pd

new_list = []
for webtoon in webtoonList:
    for webtoon_item in webtoon.get("webtoon_items"):
        new_dict = {
            "dayOfTheWeek": webtoon.get("dayOfTheWeek")
        }
        new_dict.update(webtoon_item)
        new_list.append(new_dict)

df = pd.DataFrame(new_list)
print(df)
# df.to_csv("webtoon.csv")
```
