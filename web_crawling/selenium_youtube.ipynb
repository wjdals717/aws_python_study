```python
# 검색어[ "아이폰15", "손흥민", "react", "야구"]
# 검색한 데이터에서 10개만 추출

[
    {
        "searchValue": "아이폰15",
        "data_list": [
            {
                "title": "",
                "hits": 10,
                "creator": "",
                "description": "",
                "url": ""
            }
        ]
    }
]
```

```python
# 크롬 드라이버d
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
driver.get("https://www.youtube.com/")

time.sleep(5)

# 원격으로 검색하기
searchValues = []
print("검색어를 입력해주세요.")
print("마지막은 엔터만 입력해주세요")
while(True):
    searchValue = input("검색어 입력: ")
    if(searchValue == ""):
        break
    searchValues.append(searchValue)

search_list = []
for searchValue in searchValues:
    searchInput = driver.find_element(by=By.CSS_SELECTOR, value="input[name='search_query']")
    searchInput.clear()
    searchInput.send_keys(searchValue)        # input을 통해 입력한 값이 검색창에 입력됨
    searchInput.send_keys(Keys.RETURN)        # 자동으로 enter가 되어 검색된 화면으로 이동
    time.sleep(2)       #렌더링된 이후에 코드를 실행해야 하므로 time설정은 중요함!
    
    # 스크롤
    scrollTop = 0
    while True:
        driver.execute_script(f"window.scrollTo(0, {scrollTop});")

        scrollTop += 500
        time.sleep(0.05)
        if scrollTop > 2500:
            break

    youtubediv = driver.find_elements(by=By.CSS_SELECTOR, value=".text-wrapper")
    data_list = []

    for div in youtubediv[:10]:
        title = div.find_element(by=By.CSS_SELECTOR, value='#video-title')
        title_text = title.text
        hits = div.find_element(by=By.CSS_SELECTOR, value='.ytd-video-meta-block').text.split("\n")[0]
        creator = div.find_element(by=By.CSS_SELECTOR, value='.ytd-channel-name').get_attribute(name="innerText").split("\n")[2].split("    ")[1]
        description = div.find_element(by=By.CSS_SELECTOR, value='.metadata-snippet-text').text
        url = title.get_attribute(name="href")

        data_list_item = {
            "title": title,
            "hits": hits,
            "creator": creator,
            "description": description,
            "url": url
        }
        data_list.append(data_list_item)
    data_list

    search_data = {
        "searchValue": searchValue,
        "data_list": data_list
    }
    search_list.append(search_data)
search_list


```
