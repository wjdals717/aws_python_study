```python
import pandas as pd

weatherDF = pd.read_csv("weather_data.csv",encoding="cp949")
weatherDF = weatherDF[["지점명", "일시", "기온(°C)"]]
weatherDF
```

```python
is_busan = weatherDF["지점명"] == "부산"
busan = weatherDF[is_busan]
temp = []
for date in busan["일시"]:
    temp.append(date.split(" ")[0])
temp_busan = []
unique_temp = set(temp)

for date in unique_temp:
    if temp.count(date) != 24:
        temp_busan.append(date)

empty_date = []
data_date = []
for date in busan["일시"]:
    for temp_date in temp_busan:
        if temp_date in date:
            empty_date.append(date.split(" ")[0])
            data_date.append(date)

empty_date = set(empty_date)
print(empty_date)
print(data_date)

full_date = []
for date in empty_date:
    for h in range(24):
        # full_date.append(date + f" {h:02d}:00")
        full_date.append(date + f" {str(h).zfill(2)}:00")

print(full_date)

full_empty_date_busan = []
for date in full_date:
    if date not in data_date:
        full_empty_date_busan.append(date)

print(full_empty_date_busan)
```

```python
time_list = []
for date in busan["일시"]:
    for temp_date in temp_busan:
        if("2020-08-26" in date):
            time_list.append(int(date.split(" ")[1].split(":")[0]))
print(time_list)

empty = []
for i in range(0, 24):
    find = False
    for j in time_list:
        if i == j:
            find = True
            break
    if find == False:
        empty.append(i)
print(empty)

empty = []
for i in range(0, 24):
    if i not in time_list:
        empty.append(i)
print(empty)
        
        

```

```python
is_seoul = weatherDF["지점명"] == "서울"
seoul = weatherDF[is_seoul]
temp = []
for date in seoul["일시"]:
    temp.append(date.split(" ")[0])
temp_seoul = []
unique_temp = set(temp)     #list를 set 자료형으로 형변환, 중복제거

for date in unique_temp:
    if temp.count(date) != 24:
        temp_seoul.append(date)     # 누락된 데이터가 있는 날짜 추가

empty_date = []
data_date = []
for date in seoul["일시"]:
    for temp_date in temp_seoul:
        if temp_date in date:
            empty_date.append(date.split(" ")[0])
            data_date.append(date)

empty_date = set(empty_date)
print(empty_date)
print(data_date)

full_date = []
for date in empty_date:
    for h in range(24):
        # full_date.append(date + f" {h:02d}:00")
        full_date.append(date + f" {str(h).zfill(2)}:00")

print(full_date)

full_empty_date_seoul = []
for date in full_date:
    if date not in data_date:
        full_empty_date_seoul.append(date)

print(full_empty_date_seoul)
```

```python
full_empty_date = list(set(full_empty_date_busan + full_empty_date_seoul))      # 중복제거(set) 후 list로 저장

for date in full_empty_date:
    seoul = seoul[seoul["일시"] != date]
    busan = busan[busan["일시"] != date]

seoul
busan

time = []
for i in range(24):
    time.append(str(i).zfill(2) + ":00")

import numpy as np

zeros = np.zeros((24, 2))
df = pd.DataFrame(zeros, index=time, columns=["seoul", "busan"])
df
```
